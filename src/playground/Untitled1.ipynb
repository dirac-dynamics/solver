```python
import networkx as nx
```

```python
G = nx.DiGraph()


G.add_node('a', demand = -5)
G.add_node('d', demand = 5)



G.add_edge('a', 'b', weight = 3, capacity = 4)
G.add_edge('a', 'c', weight = 6, capacity = 10)



G.add_edge('b', 'd', weight = 1, capacity = 9)
G.add_edge('c', 'd', weight = 2, capacity = 5)
flowDict = nx.min_cost_flow(G)
print(-)
```

```python
pos = nx.planar_layout(G)

nx.draw(G, pos)
nx.draw_networkx_edge_labels(G,pos)
nx.draw_networkx_labels(G,pos)
```

```python
G = nx.DiGraph()

G.add_node('a', demand = -1)
G.add_node('b', demand = -1)
G.add_node('c', demand = -1)

G.add_node('d', demand = 1)
G.add_node('e', demand = 1)
G.add_node('f', demand = 1)

G.add_edge('a', 'd', weight = 2, capacity = 1)
G.add_edge('a', 'e', weight = 10, capacity = 1)
G.add_edge('a', 'f', weight = 2, capacity = 1)

G.add_edge('b', 'd', weight = 1, capacity = 1)
G.add_edge('b', 'e', weight = 2, capacity = 1)
G.add_edge('b', 'f', weight = 2, capacity = 1)

G.add_edge('c', 'd', weight = 2, capacity = 1)
G.add_edge('c', 'e', weight = 2, capacity = 1)
G.add_edge('c', 'f', weight = 1, capacity = 1)

pos = nx.spring_layout(G)

nx.draw(G, pos)
nx.draw_networkx_edge_labels(G,pos)
nx.draw_networkx_labels(G,pos)

flowDict = nx.min_cost_flow(G)
```
